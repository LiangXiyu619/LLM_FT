## HTML

```python
import re
import html

def clean_html(text, isSimple = False):
    text = html.unescape(text)
    if isSimple:
        text = re.sub(r"<[^>]+>", " ", text)

        text = re.sub(r"(\*\*|__|>)", " ", text)

        text = re.sub(r"\s+", " ", text).strip()
        return text

    text = re.sub(r"<br\s*/?>", "\n", text, flags=re.IGNORECASE)
    text = re.sub(r"<li\s*>", "- ", text, flags=re.IGNORECASE)
    text = re.sub(r"</li\s*>", "\n", text, flags=re.IGNORECASE)
    text = re.sub(r"</?(b|strong)\s*>", "", text, flags=re.IGNORECASE)

    text = re.sub(r"<[^>]+>", " ", text)
    text = re.sub(r"(\*\*|__)", "", text)

    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\s+\n", "\n", text)
    text = re.sub(r"\n\s+", "\n", text)
    text = text.strip()
    return text
```

```python
cols_to_clean = ['query', 'product_title', 'product_description', 'product_bullet_point', 'product_brand', 'product_color']
for col in cols_to_clean:
    isSimple = (col in ['query', 'product_title', 'product_brand', 'product_color'])
    sample_12k[col + '_new'] = sample_12k[col].apply(lambda x: clean_html(x, isSimple))
    sample_12k[col + '_diff'] = (sample_12k[col] != sample_12k[col + '_new']).astype(int)

sample_12k
```

```python
import regex as re

# Unicode printable set
PRINTABLE = (
    r"\p{L}\p{N}\p{P}\p{S}\p{Zs}\p{M}"
)

def remove_garbage(text: str) -> str:
    text = re.sub(r"[\p{Cc}&&[^\n\t]]+", "", text)
    
    text = ''.join(re.findall(f"[{PRINTABLE}\n\t]", text))
    return text
```

# 原因生成

```python
e_reason = [
    "The item is relevant for the query, and satisfies all the query specifications.",
    "This product is an Exact match because its title and description clearly match the user query.",
    "This product exactly meets all the requirements stated in the search query. Its main details and description are a perfect match for what the user is looking for."
]
s_reason = [
    "The item is somewhat relevant, i.e., it fails to fulfill some aspects of the query but the item can be used as a functional substitute.",
    "This product is a Substitute because it offers similar functionality and can serve as an alternative to the item in the query.",
    "This product is a Substitute because, while not a perfect match, it provides a similar function and can effectively replace the item you're looking for."
]
c_reason = [
    "The item does not fulfill the query, but could be used in combination with an exact item.",
    "This product is a Complement because it is related to the queried item and is commonly used together with it.",
    "This product complements the query as it is designed to be used alongside the item being searched for."
]
i_reason = [
    "The item is irrelevant, or it fails to fulfill a central aspect of the query.",
    "This product is Irrelevant because it does not relate to the query in terms of product type or intended use.",
    "The product is Irrelevant because it does not match the core product type or intended use specified in the query."
]
```

```python
import random

def generate_reason(label):
    if label == "E":
        return random.choice(e_reason)
    elif label == "S":
        return random.choice(s_reason)
    elif label == "C":
        return random.choice(c_reason)
    elif label == 'I':
        return random.choice(i_reason)
    else:
        return ""
    
sample_12k['reason'] = sample_12k['esci_label'].apply(
        lambda x: generate_reason(x)
    )
sample_12k
```

```python
def rewrite_label(label):
    if label == "E":
        return "Exact"
    elif label == "S":
        return "Substitute"
    elif label == "C":
        return "Complement"
    elif label == 'I':
        return "Irrelevant"
    else:
        return ""
    
sample_12k['output'] = sample_12k['esci_label'].apply(
        lambda x: rewrite_label(x)
    )
sample_12k
```

```python
{
    "instruction": "Given the following e-commerce query and product information, determine whether the product is an Exact, Substitute, Complement, or Irrelevant match. Then explain the reason.",
    "input": "",
    "output": ""
}
```

```python
def build_input(row):
    def norm_text(x: str) -> str:
        x = x.strip()
        if x == "":
            return "Not provided."
        return x

    input_text = (
        f"Query:\n{norm_text(row.get('query'))}\n\n"
        f"Product Locale:\n{norm_text(row.get('product_locale'))}\n\n"
        f"Product Title:\n{norm_text(row.get('product_title'))}\n\n"
        f"Product Description:\n{norm_text(row.get('product_description'))}\n\n"
        f"Bullet Points:\n{norm_text(row.get('product_bullet_point'))}\n\n"
        f"Brand:\n{norm_text(row.get('product_brand'))}\n\n"
        f"Color:\n{norm_text(row.get('product_color'))}"
    )

    return input_text
```

```python
sample_12k["input"] = sample_12k.apply(
    lambda r: build_input(r.to_dict()),
    axis=1
)

sample_12k
```

```python
import re
import html
def _clean_html(text, isSimple = False):
    text = html.unescape(text)
    if isSimple:
        text = re.sub(r"<[^>]+>", " ", text)
        text = re.sub(r"(\*\*|__|>)", " ", text)
        text = re.sub(r"\s+", " ", text).strip()
        return text
    text = re.sub(r"<br\s*/?>", "\n", text, flags=re.IGNORECASE)
    text = re.sub(r"<li\s*>", "- ", text, flags=re.IGNORECASE)
    text = re.sub(r"</li\s*>", "\n", text, flags=re.IGNORECASE)
    text = re.sub(r"</?(b|strong)\s*>", "", text, flags=re.IGNORECASE)
    text = re.sub(r"<[^>]+>", " ", text)
    text = re.sub(r"(\*\*|__)", "", text)
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\s+\n", "\n", text)
    text = re.sub(r"\n\s+", "\n", text)
    text = text.strip()
    return text

import regex as re
PRINTABLE = (
    r"\p{L}\p{N}\p{P}\p{S}\p{Zs}\p{M}"
)
def _remove_garbage(text: str) -> str:
    text = re.sub(r"[\p{Cc}&&[^\n\t]]+", "", text)
    
    text = ''.join(re.findall(f"[{PRINTABLE}\n\t]", text))
    return text
```

```python
def _build_input(row):
    def norm_text(x: str) -> str:
        x = x.strip()
        if x == "":
            return "Not provided."
        return x
    input_text = (
        f"Query:\n{norm_text(row.get('query'))}\n\n"
        f"Product Locale:\n{norm_text(row.get('product_locale'))}\n\n"
        f"Product Title:\n{norm_text(row.get('product_title'))}\n\n"
        f"Product Description:\n{norm_text(row.get('product_description'))}\n\n"
        f"Bullet Points:\n{norm_text(row.get('product_bullet_point'))}\n\n"
        f"Brand:\n{norm_text(row.get('product_brand'))}\n\n"
        f"Color:\n{norm_text(row.get('product_color'))}"
    )
    return input_text
```

```python
def preprocessing(df):
    for col in cols_to_clean:
        df[col] = df[col].apply(lambda x: _clean_none(x))

    for col in cols_to_clean:
        df[col] = df[col].apply(lambda x: _clean_nfkc(x))

    for col in cols_to_clean:
        isSimple = (col in ['query', 'product_title', 'product_brand', 'product_color'])
        df[col] = df[col].apply(lambda x: _clean_html(x, isSimple))

    for col in cols_to_clean:
        df[col] = df[col].apply(lambda x: _remove_garbage(x))

    for col in ['product_description', 'product_bullet_point']:
        df[col] = df[col].apply(
            lambda x: _truncate_by_tokens(x)
        )
    
    df["input"] = df.apply(
        lambda r: _build_input(r.to_dict()),
        axis=1
    )

    return df
```
